#Refactorizando el loop for:

```
def func_e_to_2(n):
    x = 2
    e_to_2 = 0
    for i in range(n):
        e_to_2 += x**i/math.factorial(i)
        
    return(e_to_2)
print(func_e_to_2(10))   
print(math.exp(2))
```

```
def func_sin_to_2(n):
    sin_to_2 = 0
    sign = 1 
    x = 2
    for i in range(n):
        
        sin_to_2 += (x**(2*i+1)/math.factorial(2*i+1))*sign
        sign *= -1
    return(sin_to_2)
print(func_sin_to_2(10))
print(math.sin(2))
```

```
def func_cos_to_2(n):
    cos_to_2 = 0
    sign = 1 
    x = 2
    for i in range(n):
        
        cos_to_2 += (x**(2*i)/math.factorial(2*i))*sign
        sign *= -1
    return(cos_to_2)
print(func_cos_to_2(10))
print(math.cos(2))
```

```
def func_mul_inv_1_minus_x(n):
    x= 0.5
    mul_inv_1_minus_x = 0

    for i in range(n):
        mul_inv_1_minus_x += x**(i)
    return(mul_inv_1_minus_x )
print(func_mul_inv_1_minus_x(10))
print(1/(1-x))
```

```
def func_ln_1_plus_x(n):
    ln_1_plus_x = 0
    sign = 1
    x = 0.5
    for i in range(1,n):
        ln_1_plus_x += (x**i/i)*sign
        sign *= -1
    return(ln_1_plus_x)
print(func_ln_1_plus_x(10))
print(math.log(1 + x))
```

```
def func_ln_x(n):
    ln_x = 0
    sign = 1
    x = 0.5
    for i in range(1,n):
        ln_x += ((x-1)**i/i)*sign
        sign *= -1
    return(ln_x)
print(func_ln_x(10))
print(math.log(x))
```

#Recibiendo x como parámetro de nuestras funciones:

```
def func_e(x,n):
    e = 0
    for i in range(n):
        e += x**i/math.factorial(i)
        
    return(e)
print(func_e(2,10))   
print(math.exp(2))
```

```
def func_sin(x,n):
    sin = 0
    sign = 1 
    for i in range(n):
        
        sin += (x**(2*i+1)/math.factorial(2*i+1))*sign
        sign *= -1
    return(sin)
print(func_sin(2,10))
print(math.sin(2))
```

```
def func_cos(x,n):
    cos = 0
    sign = 1 
    for i in range(n):
        
        cos += (x**(2*i)/math.factorial(2*i))*sign
        sign *= -1
    return(cos)
print(func_cos(2,10))
print(math.cos(2))
```

```
def func_mul_inv_1_minus_x2(x,n):
    mul_inv_1_minus_x = 0

    for i in range(n):
        mul_inv_1_minus_x += x**(i)
    return(mul_inv_1_minus_x )
print(func_mul_inv_1_minus_x2(0.5,10))
print(1/(1-0.5))
```

```
def func_ln_1_plus_x2(x,n):
    ln_1_plus_x = 0
    sign = 1
    for i in range(1,n):
        ln_1_plus_x += (x**i/i)*sign
        sign *= -1
    return(ln_1_plus_x)
print(func_ln_1_plus_x2(0.5,10))
print(math.log(1 + 0.5))
```

```
def func_ln_x2(x,n):
    ln_x = 0
    sign = 1
    for i in range(1,n):
        ln_x += ((x-1)**i/i)*sign
        sign *= -1
    return(ln_x)
print(func_ln_x2(0.5,10))
print(math.log(0.5))
```

```
from sympy import series, Symbol, Function
from sympy.functions import sin, cos, exp, ln
from sympy.plotting import plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = 13,10
plt.rcParams['lines.linewidth'] = 2
```

```
class f(Function):
    @classmethod
    def eval(cls, x):
        return 1/(1-x)    

class g(Function):
    @classmethod
    def eval(cls, x):
        return ln(x+1)
```

```
x = Symbol('x')
```

```
def taylor(function, x0, n):
    """
    Parameter "function" is our function which we want to approximate
    "x0" is the point where to approximate
    "n" is the order of approximation
    """
    return function.series(x,x0,n).removeO()
```

```
print('sin(x) =', taylor(sin(x), 0, 4))

print('cos(x) =', taylor(cos(x), 0, 4))

print('e(x) =', taylor(exp(x), 0, 4))

print('f(x) = 1/(1-x) =', taylor(f(x), 0, 4))

print('g(x) = ln(x+1) =', taylor(g(x), 0, 4))

print('ln(x) =', taylor(ln(x), 0, 4))
```

```
print('sin(1) =', taylor(sin(x), 0, 4).subs(x,1))

print('cos(1) =', taylor(cos(x), 0, 4).subs(x,1))

print('e(1) =', taylor(exp(x), 0, 4).subs(x,1))

print('f(1) =', taylor(f(x), 0, 4).subs(x,1))

print('g(1) =', taylor(g(x), 0, 4).subs(x,1))

print('ln(1) =', taylor(ln(x), 0, 4).subs(x,1))
```

```
# This will plot sine and its Taylor approximations

p = plot(sin(x),taylor(sin(x),0,1),taylor(sin(x),0,3),taylor(sin(x),0,5),
         (x,-3.5,3.5),legend=True, show=False)

p[0].line_color = 'blue'
p[1].line_color = 'green'
p[2].line_color = 'firebrick'
p[3].line_color = 'black'
p.title = 'Taylor Series Expansion for Sine'
p.show()
```
